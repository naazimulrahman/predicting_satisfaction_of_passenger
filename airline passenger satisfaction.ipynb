```python
data['Arrival Delay in Minutes'] = data['Arrival Delay in Minutes'].mean()
```

```python
data.drop(columns=['Unnamed: 0','id'], inplace=True)
```

```python
x = data.iloc[ : , :-1]
y = data.iloc[ : , -1]
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
y = label.fit_transform(y)
y
```

```python
categories = x.select_dtypes(include=['object']).columns.tolist()
categories
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
transformer = ColumnTransformer([
    ('encoder', OneHotEncoder(handle_unknown='ignore', drop="first"), categories),
], remainder=MinMaxScaler())
x = transformer.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
import tensorflow
from tensorflow import keras
from keras.regularizers import l2
from keras.models import Sequential
from keras.layers import Dense, Input ,Dropout ,BatchNormalization
```

```python
model = Sequential()
model.add(Input(shape=(x_train.shape[1],)))
model.add(BatchNormalization())
model.add(Dense(64, activation="relu",  kernel_initializer="glorot_uniform",kernel_regularizer=l2(0.03),name="layer1"))
model.add(Dropout(rate=0.1))
model.add(Dense(32, activation="relu", kernel_initializer="glorot_uniform", name="layer2"))
model.add(Dense(1, activation="sigmoid", name="layer3"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(x_train, y_train, validation_data =(x_test, y_test), batch_size = 32, epochs = 10)
```

```python
metrics = pd.DataFrame(history.history)
metrics[["loss","val_loss"]].plot();
metrics[["accuracy","val_accuracy"]].plot();
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
    """

    Args:
      hp:

    Returns:

    """
    model = Sequential()
    counter =0
    for i in range(hp.Int('num_layers', min_value = 2, max_value =20)):
      if counter == 0:

        model.add(
            Dense(
                units = hp.Int('units_' + str(i), min_value = 32, max_value =512, step = 32),
                activation = hp.choice('activation_' + str(i), values = ['relu', 'tanh','sigmoid']),
                input_dim = x_train.shape[1]
            )
        )
      else:
        model.add(
            Dense(
                units = hp.Int('units_' + str(i), min_value = 32, max_value =512, step = 32),
                activation = hp.choice('activation_' + str(i), values = ['relu', 'tanh','sigmoid'])
            )
        )
      counter+=1
    model.add(Dense(1, activation='sigmoid'))
    # The following line was indented too far, causing the error.
    model.compile(
        optimizer=hp.Choice('optimizer', values = ['adam', 'rmsprop', 'sgd']),
        loss=hp.Choice('loss', values = ['binary_crossentropy', 'mean_squared_error']),
        metrics=['accuracy']
    )
    return model
```

```python
tuner=kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='kt_dir',
    project_name='nazoo'
)
```

```python
tuner.search(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```
